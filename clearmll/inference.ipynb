```python
import time
import requests
import pandas as pd
from fitness_package.preprocess import get_data
```

```python
endpoint = "http://localhost:8080/serve/xgb_fitness_class"
```

```python
data_path = get_data(
    'fitness_class_inference_data', 
    r'^encoded.*',
)
data = pd.read_csv(data_path)
```

```python
data.head()
```

```python
for i in range(10):
    payload =  data.iloc[i, :-1].to_dict()
    try:
        response = requests.post(endpoint, json=payload)
    except:
        pass
    time.sleep(0.2)
    print(payload)
    print(response.text)
    print('\n')
```
